# Matplot-Prac

### Imports

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

### Arrays

```python
x = np.linspace(0,7,15)
y = x ** 2
```

### Object Orientated Sub Plotting

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
```

```python
axes[0].plot(x, y, 'r')
axes[0].set_title("Left Plot")

axes[1].plot(y, x, 'b')
axes[1].set_title("Right Plot")

fig
```

### Saving figure to an image file

```python
fig.savefig("savefig-test.png",dpi=200)
```

### Labels, Linewidth, Color and Legend

```python
plt.plot(x, y, label = "First label", linewidth = 2, color = 'blue',marker='o')
plt.plot(y, x, label = "Second label", linewidth = 3, color = 'orange', linestyle= ':')
plt.legend(loc=10)
```

```python
plt.scatter(x, y, label = "First label", color = 'blue')
plt.scatter(y, x, label = "Second label", color = 'orange')
plt.legend()
```
