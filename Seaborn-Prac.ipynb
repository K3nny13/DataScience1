```python
import seaborn as sns

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

```python
sns.distplot(tips['total_bill'],bins=30)
```

```python
sns.pairplot(tips,hue="sex")
```

## KDE : Kernel Density Estimation

```python
sns.kdeplot(tips["total_bill"])
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.swarmplot(x='sex',y='total_bill',data=tips)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='blue')
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='violin')
```

## Correlation

```python
tc = tips.corr()
```

```python
sns.heatmap(tc,annot=True)
```

```python
flights = sns.load_dataset('flights')
fp = flights.pivot_table(index='month',columns='year',values='passengers')
sns.heatmap(fp,linewidth=3)
```

```python
iris = sns.load_dataset('iris')

k = sns.PairGrid(iris)
k.map_diag(sns.distplot)
k.map_upper(sns.scatterplot)
k.map_lower(sns.kdeplot)
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex',markers=['o','v'])
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',col='sex')
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',col='sex',row='time')
```
